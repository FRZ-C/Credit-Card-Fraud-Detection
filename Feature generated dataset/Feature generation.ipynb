**Single address but multiple customers flag (address_multiple_customer_flag)**
---



```python
# Train dataset
# Ensure that cc_num, first, last, and zip are in the correct format (if they are not already strings)
df['cc_num'] = df['cc_num'].astype(str)
df['first'] = df['first'].astype(str)
df['last'] = df['last'].astype(str)
df['zip'] = df['zip'].astype(str)

# Group by first name, last name, and zip, then count unique credit cards
cc_counts = df.groupby(['first', 'last', 'zip'])['cc_num'].nunique().reset_index(name='cc_count')

# Merge this count back with the original DataFrame to get the Cc_count for each row
df = pd.merge(df, cc_counts, on=['first', 'last', 'zip'], how='left')

# To check if one address (combination of street, city, state, zip) belongs to more than one customer
# Create a unique identifier for each address
df['address_id'] = df['street'] + ', ' + df['city'] + ', ' + df['state'] + ', ' + df['zip']

# Count the number of unique customers (first name, last name) per address
address_counts = df.groupby('address_id')['first', 'last'].apply(lambda x: len(x.drop_duplicates())).reset_index(name='unique_customer_address')

# Map this count to a new column, where 1 means the address belongs to more than one customer, 0 otherwise
df = pd.merge(df, address_counts, on='address_id', how='left')
df['address_multiple_customer_flag'] = df['unique_customer_address'].apply(lambda x: 1 if x > 1 else 0)
```

```python
# Test dataset
# Ensure that cc_num, first, last, and zip are in the correct format (if they are not already strings)
test_df['cc_num'] = test_df['cc_num'].astype(str)
test_df['first'] = test_df['first'].astype(str)
test_df['last'] = test_df['last'].astype(str)
test_df['zip'] = test_df['zip'].astype(str)

# Group by first name, last name, and zip, then count unique credit cards
cc_counts_test = test_df.groupby(['first', 'last', 'zip'])['cc_num'].nunique().reset_index(name='cc_count')

# Merge this count back with the original DataFrame to get the Cc_count for each row
test_df = pd.merge(test_df, cc_counts_test, on=['first', 'last', 'zip'], how='left')

# Create a unique identifier for each address
test_df['address_id'] = test_df['street'] + ', ' + test_df['city'] + ', ' + test_df['state'] + ', ' + test_df['zip']

# Count the number of unique customers (first name, last name) per address
# Adjusting the lambda function to work with the updated pandas version
address_counts_test = test_df.groupby('address_id')['first', 'last'].apply(lambda x: len(x.drop_duplicates())).reset_index(name='unique_customer_address')
# Map this count to a new column, where 1 means the address belongs to more than one customer, 0 otherwise
test_df = pd.merge(test_df, address_counts_test, on='address_id', how='left')
test_df['address_multiple_customer_flag'] = test_df['unique_customer_address'].apply(lambda x: 1 if x > 1 else 0)
```

**Distance between customer and merchant (transaction_distance)**
---



From EDA it looks like there is not much siginificance. But, the feature can be calculated and fed to see if it contributes anything.

```python
# Train dataset
def haversine(lat1, lon1, lat2, lon2):
    """
    Calculate the great circle distance in kilometers between two points
    on the earth (specified in decimal degrees)
    """
    # Convert decimal degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    r = 6371  # Radius of earth in kilometers
    return c * r

# Calculate distance and add it as a new column
df['Latlong_distance'] = df.apply(lambda x: haversine(x['lat'], x['long'], x['merch_lat'], x['merch_long']), axis=1)

# Define distance thresholds for 'short', 'medium', and 'far'
short_distance_threshold = 20  # distances up to 20 km
medium_distance_threshold = 50  # distances between 20 km and 50 km

# Categorize the distances into 'short', 'medium', 'far' with numerical encoding
def categorize_distance(distance):
    if distance <= short_distance_threshold:
        return 0  # 'short'
    elif distance <= medium_distance_threshold:
        return 1  # 'medium'
    else:
        return 2  # 'far'

df['transaction_distance'] = df['Latlong_distance'].apply(categorize_distance)
```

```python
# Test dataset
# Calculate distance for test_df and add it as a new column
test_df['Latlong_distance'] = test_df.apply(lambda x: haversine(x['lat'], x['long'], x['merch_lat'], x['merch_long']), axis=1)

# Define distance thresholds for 'short', 'medium', and 'far'
short_distance_threshold = 20  # distances up to 20 km
medium_distance_threshold = 50  # distances between 20 km and 50 km

test_df['transaction_distance'] = test_df['Latlong_distance'].apply(categorize_distance)
```

**Age and fraud likelihood (fraud_likelihood_age)**
---

```python
# Train dataset
# Calculate age as before
df['dob'] = pd.to_datetime(df['dob'])
df['age'] = df['trans_date_trans_time'].dt.year - df['dob'].dt.year
df.loc[df['trans_date_trans_time'].dt.month < df['dob'].dt.month, 'age'] -= 1
df.loc[(df['trans_date_trans_time'].dt.month == df['dob'].dt.month) &
       (df['trans_date_trans_time'].dt.day < df['dob'].dt.day), 'age'] -= 1

# Define age range for 'highly likely' classification
age_lower_bound = 30
age_upper_bound = 70

# Classify age groups into 'highly likely' (1) or 'less likely' (0)
df['fraud_likelihood_age'] = df['age'].apply(lambda x: 1 if age_lower_bound <= x <= age_upper_bound else 0)
```

```python
# Test dataset
import pandas as pd

# Convert 'dob' to datetime format in test_df
test_df['dob'] = pd.to_datetime(test_df['dob'])

# Calculate age based on 'trans_date_trans_time' and 'dob'
test_df['age'] = test_df['trans_date_trans_time'].dt.year - test_df['dob'].dt.year

# Adjust for cases where the birthday hasn't occurred yet this year
test_df.loc[test_df['trans_date_trans_time'].dt.month < test_df['dob'].dt.month, 'age'] -= 1
test_df.loc[(test_df['trans_date_trans_time'].dt.month == test_df['dob'].dt.month) &
            (test_df['trans_date_trans_time'].dt.day < test_df['dob'].dt.day), 'age'] -= 1

# Define age range for 'highly likely' classification (same as used in training dataset)
age_lower_bound = 30
age_upper_bound = 70

# Classify age groups into 'highly likely' (1) or 'less likely' (0) for fraud likelihood
test_df['fraud_likelihood_age'] = test_df['age'].apply(lambda x: 1 if age_lower_bound <= x <= age_upper_bound else 0)
```

**Large orders by multiple cards made by one person (multiple_cards_large_orders)**
---



```python
# Train dataset
# Calculate the average amount spent for each category and merchant
average_spend = df.groupby(['category', 'merchant'])['amt'].mean().reset_index(name='avg_amt')

# Merge this average back into the original df to compare individual transactions
df = df.merge(average_spend, on=['category', 'merchant'], how='left')

# Flag orders that are larger than the average spend for their specific category and merchant
df['large_order'] = (df['amt'] > df['avg_amt']).astype(int)

# Group by person (first name, last name, and zip code) and count distinct cards used for large orders
large_orders_cards = df[df['large_order'] == 1].groupby(['first', 'last', 'zip'])['cc_num'].nunique().reset_index(name='distinct_cards')

# Create a binary feature that flags individuals using more than 3 distinct cards for large orders
large_orders_cards['multiple_cards_large_orders'] = (large_orders_cards['distinct_cards'] > 3).astype(int)

# Merge this binary feature back into the original DataFrame
# First, create an identifier for each person in both DataFrames for merging
df['person_id'] = df[['first', 'last', 'zip']].agg('_'.join, axis=1)
large_orders_cards['person_id'] = large_orders_cards[['first', 'last', 'zip']].agg('_'.join, axis=1)

# Merge on the 'person_id' column to include the multiple_cards_large_orders flag in the original DataFrame
df = df.merge(large_orders_cards[['person_id', 'multiple_cards_large_orders']], on='person_id', how='left')

# Replace NaN with 0 for those without large orders or less than 3 distinct cards
df['multiple_cards_large_orders'] = df['multiple_cards_large_orders'].fillna(0)
```

```python
# Test dataset
# Assume 'average_spend' was calculated from the training dataset 'df' as described earlier
# and 'test_df' has been loaded and prepared up to this point.

# Merge the average spend information with test_df based on category and merchant
test_df = test_df.merge(average_spend, on=['category', 'merchant'], how='left')

# Flag orders that are larger than the average spend for their specific category and merchant in test_df
test_df['large_order'] = (test_df['amt'] > test_df['avg_amt']).astype(int)

# Calculate distinct cards used for large orders by person in test_df
large_orders_cards_test = test_df[test_df['large_order'] == 1].groupby(['first', 'last', 'zip'])['cc_num'].nunique().reset_index(name='distinct_cards')

# Create a binary feature for using more than 3 distinct cards for large orders in test_df
large_orders_cards_test['multiple_cards_large_orders'] = (large_orders_cards_test['distinct_cards'] > 3).astype(int)

# Create an identifier for each person in both the test_df and large_orders_cards_test DataFrames for merging
test_df['person_id'] = test_df[['first', 'last', 'zip']].agg('_'.join, axis=1)
large_orders_cards_test['person_id'] = large_orders_cards_test[['first', 'last', 'zip']].agg('_'.join, axis=1)

# Merge this binary feature back into the original test_df
test_df = test_df.merge(large_orders_cards_test[['person_id', 'multiple_cards_large_orders']], on='person_id', how='left', sort=False)

# Replace NaN with 0 for those without large orders or less than 4 distinct cards in test_df
test_df['multiple_cards_large_orders'] = test_df['multiple_cards_large_orders'].fillna(0).astype(int)
```
